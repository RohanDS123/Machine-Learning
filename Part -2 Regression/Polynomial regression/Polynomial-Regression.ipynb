# Polynomial Regression

# 1.  Importing the libraries.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

# 2.  Importing the dataset 

```python
df = pd.read_csv("Position_Salaries.csv")
```

```python
df
```

```python
X = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

# 3.Training the linear regression on the whole datset - 

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X,y)
```

# 4.Training the polynomial regression model on the whole dataset - 

```python
from sklearn.preprocessing import PolynomialFeatures
regressor2 = PolynomialFeatures(degree = 4)
apply = regressor2.fit_transform(X)
new_one = LinearRegression()
new_one.fit(apply,y)
```

# 5. Visualize the linear regression result - 

```python
plt.scatter(X,y,color = "red")
plt.plot(X,regressor.predict(X),color = "blue")
plt.xlabel("Level",fontsize = 17,color = "red")
plt.ylabel("Salaries",fontsize = 17,color = "red")
plt.title("Polynomial Model",fontsize = 17,color = "red")
plt.show()
```

# 6. Visualize the polynomial regression model - 

```python
plt.scatter(X,y,color = "red")
plt.plot(X,new_one.predict(regressor2.fit_transform(X)),color = "blue")
plt.xlabel("Level",fontsize = 17,color = "red")
plt.ylabel("Salaries",fontsize = 17,color = "red")
plt.title("Polynomial Model",fontsize = 17,color = "red")
plt.show()
```

# 7.Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, new_one.predict(regressor2.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

# Predicting the new result with linear regression.

```python
regressor.predict([[6]])
```

# Predicting the new result with polynomail regression.

```python
new_one.predict(regressor2.fit_transform([[6]]))
```
