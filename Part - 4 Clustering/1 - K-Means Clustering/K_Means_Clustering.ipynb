#**Clusterig -**
**Clustering in machine learning is a process of grouping of the similar data points together based on certain features,without prior knowledge of their labels. The main aim of clustering is finding the best structure and patters for future analysis and classification task.**

**Clustering is a unsupervise machine learning technique in which we don't need to  labeled data we use of unlabeled data.**

# **Real life example -**
**We use of clustering in machine learning such as customer segmentation for decide the best decision for our business growth.**

# **The Elbow Method -**
**The Elbow method is help use to decide how many clusters we want.**

#**Importing the libraries:**

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

#**Importing the dataset:**

```python
data = pd.read_csv("Mall_Customers.csv")
x = data.iloc[:,[3,4]].values  # Here i'm selecting only 3rd and 4rth column becase it is best for costomer segmentation.
```

```python
x
```

# **Using the Elbow method find the optimal number of clusters:**

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters = i,init = "k-means++",random_state = 42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)  # Here kmeans.inertia_ helps us to identify the optimal no.of clusters by using Elbow method
plt.plot(range(1,11),wcss,color = "b")
plt.title("The Elbow method",fontsize = 17,color = 'k')
plt.xlabel("Range",fontsize = 16,color= 'k')
plt.ylabel("Wcss",fontsize = 16,color = 'k')
plt.show()
```

# **Training the kmeans model on the dataset:**

```python
kmeans = KMeans(n_clusters = 5,init = "k-means++",random_state = 42)
y_kmeans = kmeans.fit_predict(x)
```

```python
y_kmeans
```

# **Visualize the result:**

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(x[y_kmeans == 3, 0], x[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(x[y_kmeans == 4, 0], x[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
