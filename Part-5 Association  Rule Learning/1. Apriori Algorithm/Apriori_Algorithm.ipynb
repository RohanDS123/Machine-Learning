# **Apriori Algorithm -**
**Apriori algorithm is a method of machine learning which is used to find the relationship between variables and large datasets:**

# **Importing the libraries:**

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

# **In this algorithm we will not use of scikit learn library but we will use of Apyori library**

# **Data preprocessing:**

```python
data = pd.read_csv("Market_Basket_Optimisation.csv",header = None)
new_one = []
for i in range(0,7501):
  new_one.append([str(data.values[i,j]) for j in range(0,20)])
```

**This code is converting pandas csv file into python file it's make easier to learn Association rule mining:**

```python
data
```

# **Training the apriori model on the dataset:**

```python
from apyori import apriori
rules = apriori(transactions = new_one,min_support = 0.003,min_confidence = 0.2,min_lift = 3,min_length = 2,max_length = 2)
```

**|min_length=2: This specifies the minimum number of items in each rule. Here, it's set to 2, meaning we're looking for association rules with two items.
max_length=2: This specifies the maximum number of items in each rule. Again, it's set to 2, meaning we're only interested in association rules with exactly two items on both sides.**
